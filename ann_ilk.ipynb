```
#Gereken kutuphaneler yuklenmeli
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn.model_selection import train_test_split  #Rastgele Train Test Bolmek icin
from sklearn.preprocessing import MinMaxScaler        #Min-Max normalization yapmak icin
import numpy

numpy.random.seed(7)
#veriseti yukleme
dataset=numpy.loadtxt("Final_YeniVeriler.csv", delimiter=",")

#datalar 0:5 arasi featurelar alindi
datalar=dataset[:,0:5]
datalar=MinMaxScaler().fit_transform(datalar)
#Veriler [0 1] araligina normalize edildi

etiketler=dataset[:,5]
#etiketler 5.indis class (outcome) alindi
etiketler=np_utils.to_categorical(etiketler, 2)
# Class isimleri 0,1 iken 2 giris icin ayarlandi 
# 1= [1,0], 2=[0,1], ...

# Splitting data
X_train, X_test, Y_train, Y_test = train_test_split(datalar, etiketler, test_size=0.2, random_state=0)
#X_train: Train Datalari
#X_test:  Test Dataları
#Y_train: Train Etiketleri
#Y_test:  Test Etiketleri
```

# **MLP YAPISI KURULUMU**

```
import random

traning_func = ['tanh' ]  #,'linear','tanh',  'softmax', 'sigmoid'
hidden2=['relu', 'linear','tanh', 'softmax', 'sigmoid' ]  #
hidden3=['relu', 'linear','tanh', 'softmax','sigmoid'] 
transfer_func = ['relu', 'linear','tanh',  'softmax', 'sigmoid']
sayac = 1
for i in traning_func: #hidden 1
  for x in hidden2:  #hidden 2 
    for y in hidden3:  # hidden 3 
      for j in transfer_func:   # output layer
        for k in range(0, 5):   # her modeli 5er kere çalıştır
          a = random.randint(80,120)  #hidden 1 
          b = random.randint(90,200)  #hidden 2 
          c = random.randint(90,200)  #hidden 3 
          d = random.randint(200,800)  # epoch

          model=Sequential()
          #5 feature girisli oldugu icin input_dim=5 secildi
          model.add(Dense(a, input_dim=5, activation=i)) # Hidden Layer 1 (25 noron)
          #hidden layer 2 ye input_dim yazilmaz.
          model.add(Dense(b, activation=x))              # Hidden Layer 2  (100 noron)
          model.add(Dense(c, activation=y))              # Hidden Layer 2  (100 noron)
          #Cikis 2 adet oldugu icin oldugu icin Dense 7 output noron baglandi
          model.add(Dense(2, activation=j))            #Output Layer 2çıkışlı
          #activation=sigmoid, relu, linear, tanh, softmax

          #Modeli Derleme
          model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
          #loss=KLD, mae, mse, mean_squared_error, cosine_proximity, binary_crossentropy, poisson, categorical_crossentropy
          #optimizer: adam, RMSprop, SGD

          #batch_size: Egitim sirasinda verisetinden ornekler kacar kacar alinacak
          model.fit(X_train, Y_train, epochs=d, batch_size=10, verbose=0)
          #verbose=0 -> Eğitimi adim adim gostermez sadece son epoch getirir
          basarim=model.evaluate(X_train, Y_train)
          _, test_acc = model.evaluate(X_test, Y_test, verbose=0)

          print("\nSayac=%d,Hidden1 = %d, Hidden2= %d, Hidden3= %d, epochs = %d, Traning Function = %s_%s_%s, Transfer Function = %s, loss = binary_crossentropy" % (sayac, a, b, c, d, i, x, y, j))  #     
          print("\nTrain %s : %.2f%% " % (model.metrics_names[1], basarim[1]*100))
          print('\nTest Accuracy: %.2f%%' % (test_acc * 100.0))
          model.save(str(sayac)+"_"+"yeni_"+str(i)+"_"+str(x)+"_"+str(y)+"_"+str(round(test_acc * 100.0, 2))+".h5")
          #model.save("yeniveri_"+str(i)+"_"+str(sayac)+"_"+str(round(test_acc * 100.0, 2))+".h5")
          sayac += 1
```

```
from sklearn.metrics import classification_report,confusion_matrix

_, test_acc = model.evaluate(X_test, Y_test, verbose=0)
print('>Test Accuracy: %.2f%%' % (test_acc * 100.0))


#yPredicted = model.predict(X_test)           #Direk cikista alinan degerleri verir
#yPredicted = model.predict_proba(X_test)     #Olasiliksal olarak cikti verir
yPredicted = model.predict_classes(X_test)
Y_Test_actual=numpy.argmax(Y_test,axis=1)

print(classification_report(y_true=Y_Test_actual, y_pred=yPredicted))
print(confusion_matrix(y_true=Y_Test_actual, y_pred=yPredicted))
```
